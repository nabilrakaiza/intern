```python
import json
from typing import Optional, Literal
from pydantic import BaseModel, Field
from openai_helpers import call_llm

class ArticleTagsRetriever(BaseModel):
    """
    Retrieve tag(s) from given article
    """
    tags: str = Field(description="The tag(s) of the article")

def get_tags_from_article(article : str):
    extracted_tags = call_llm(
        model=ArticleTagsRetriever,
        prompt_template=gpt_system_prompt,
        params={"article":article}
    )
    return extracted_tags.tags.split(",")

gpt_system_prompt = """
You are an expert tag(s) extraction. Your task is to identify and extract up to 10 relevant tag(s) (may be less) from the provided article. 
These tags should accurately reflect the main topics, entities, and keywords present in the article.

--
Article Content : 
{article}
--

**Guidelines:**
* **Relevance:** All extracted tags must be directly related to the content of the article.
* **Specificity:** Only put the specific terms, and omit the general ones.
* **Format:** Return the tags as a single comma-separated string (CSV format), with NO extra spaces.
"""

with open('presidential.issues.json', 'r') as file:
    datajson = json.load(file)

important_field = ['title',
 'main_problem',
 'summary',
 'keypoints',
 'government_actions_already_taken',
 'government_institutions_involved',
 'non_government_institutions_involved',
 'executive_comparing_summary_for_government_and_non_government_institutions',
 'statistics',
 'subtitle',
 'content',
 'issue_category',
 'gepolitic_aspects',
 'peoples_involved']

article_data = []
n = 1

for data in datajson:

    article = ""

    for field in important_field:
        article += str(f"{field}: {data[field]}\n\n")
    
    tags = get_tags_from_article(article)

    dict_res = {}
    dict_res["id"] = n
    dict_res["title"] = data["title"]
    dict_res["tags"] = ",".join(tags)

    n+=1

    article_data.append(dict_res)
```

```python
import os
from langchain.chains import RetrievalQA
from langchain_openai import ChatOpenAI as ChatOpenAIRAG
from langchain_openai import OpenAIEmbeddings
from langchain_community.vectorstores import FAISS
from langchain_core.documents import Document # Explicitly import Document

# Assume these are defined elsewhere, similar to your original code
# For demonstration, I'll provide a dummy implementation.
# In a real scenario, these would load/process your news articles.

def get_all_news_articles() -> list[dict]:
    """
    Simulates fetching all news articles.
    Each article is a dictionary with at least 'content' and 'id'.
    """
    return [
        {"id": "news_1", "title": "Local Bakery Wins Award", "content": "A small bakery in the town of Springfield has won the prestigious 'Golden Crumb' award for its sourdough bread."},
        {"id": "news_2", "title": "Tech Giant Announces New AI Product", "content": "InnovateCo, a leading technology company, unveiled its groundbreaking AI-powered personal assistant, promising to revolutionize daily tasks."},
        {"id": "news_3", "title": "Sports Team Clinches Championship", "content": "The Springfield Hornets secured their first basketball championship in decades, with a thrilling victory in the final game."},
        {"id": "news_4", "title": "Global Climate Summit Concludes", "content": "World leaders met to discuss climate change policies, reaching agreements on reducing carbon emissions and investing in renewable energy."},
        {"id": "news_5", "title": "New Study on Sleep Habits Released", "content": "Researchers have published findings indicating a strong correlation between regular sleep patterns and improved cognitive function."},
        {"id": "news_6", "title": "Financial Markets See Volatility", "content": "Stock markets experienced significant fluctuations this week, driven by geopolitical events and inflation concerns."},
        {"id": "news_7", "title": "Community Project to Revitalize Park", "content": "Local volunteers have launched an initiative to clean up and renovate the central park, aiming to create a better public space."},
        {"id": "news_8", "title": "Another article about AI breakthroughs", "content": "A startup has developed a novel AI algorithm that significantly improves image recognition accuracy."},
    ]

def filter_valid_news_articles(
    all_articles: list[dict], list_of_news_ids_to_consider: list[str]
) -> list[dict]:
    """
    Simulates filtering news articles based on a list of valid IDs.
    """
    if not list_of_news_ids_to_consider:
        return all_articles # If no specific IDs are given, consider all
    
    valid_articles = []
    valid_ids_set = set(list_of_news_ids_to_consider)
    for article in all_articles:
        if article.get("id") in valid_ids_set:
            valid_articles.append(article)
    return valid_articles

def group_news_by_content(news_articles: list[dict]) -> dict[str, dict]:
    """
    Creates a mapping from news article content to the full article dictionary.
    This is analogous to your `group_template_id_by_ratio`.
    """
    grouped_articles = {}
    for article in news_articles:
        # Using content as key, assuming content is unique enough for this purpose
        # In a real system, you might use a stable hash of the content or an explicit ID
        grouped_articles[article["content"]] = article
    return grouped_articles

def RAG_find_similar_news_article(
    user_input_news_content: str, list_of_news_ids_to_consider: list[str] = None
) -> tuple[str, dict]:
    """
    Finds the news article most similar to the user's input news content
    from a list of available news articles.

    Args:
        user_input_news_content (str): The content of the news article
                                       for which to find similar articles.
        list_of_news_ids_to_consider (list[str], optional): A list of news article IDs
                                                            to limit the search.
                                                            If None, all available news
                                                            articles are considered.

    Returns:
        tuple[str, dict]: A tuple containing:
            - The content of the most similar news article found.
            - The full dictionary of the most similar news article (including ID, title, etc.).
    """
    
    # Ensure OpenAI API Key is set
    if "OPENAI_API_KEY" not in os.environ:
        raise ValueError("OPENAI_API_KEY environment variable not set.")

    embeddings = OpenAIEmbeddings(api_key=os.environ["OPENAI_API_KEY"])

    # 1. Get all news articles
    all_news_articles = get_all_news_articles()

    # 2. Filter news articles if a specific list of IDs is provided
    valid_news_articles = filter_valid_news_articles(
        all_news_articles, list_of_news_ids_to_consider
    )

    # 3. Prepare data for vector store: map content to full article data
    # We'll use the content as the text to embed and search by
    grouped_news_by_content = group_news_by_content(valid_news_articles)

    # 4. Create Langchain Document objects from news article content
    documents = [
        Document(page_content=article_content, metadata={"id": article_data["id"], "title": article_data["title"]})
        for article_content, article_data in grouped_news_by_content.items()
    ]

    # 5. Create FAISS vector store
    # This embeds the content of each news article and indexes them
    vector_store = FAISS.from_documents(documents, embeddings)

    # 6. Set up the retriever
    # We want to retrieve the single most similar news article
    retriever = vector_store.as_retriever(
        search_type="similarity", search_kwargs={"k": 1}
    )

    # 7. Initialize the Language Model for RAG (optional for pure similarity, but good for context)
    llm_rag = ChatOpenAIRAG(
        model="gpt-4o", temperature=0, api_key=os.environ["OPENAI_API_KEY"]
    )

    # 8. Create the RetrievalQA chain
    # The chain will take the user's news content, retrieve similar articles,
    # and then (optionally) use the LLM to refine the "answer" if needed.
    qa_chain = RetrievalQA.from_chain_type(
        llm=llm_rag,
        chain_type="stuff", # "stuff" means retrieved docs are put directly into the prompt
        retriever=retriever,
        return_source_documents=True, # We want to see which document was retrieved
    )

    # 9. Execute the RAG chain with the user's input news content
    result = qa_chain({"query": user_input_news_content})

    # 10. Extract the most similar news article content
    # Since k=1, result["source_documents"] will have one document
    if result["source_documents"]:
        top_doc_content = result["source_documents"][0].page_content
        top_doc_metadata = result["source_documents"][0].metadata
        
        # Retrieve the full article dictionary using the content as the key
        most_similar_article_data = grouped_news_by_content[top_doc_content]
        
        return top_doc_content, most_similar_article_data
    else:
        return "No similar news article found.", {}

# Example Usage (make sure to set your OPENAI_API_KEY environment variable)
if __name__ == "__main__":
    # Dummy API key for demonstration if not set.
    # In a real application, set it securely: export OPENAI_API_KEY="your_key_here"
    # or use a library like python-dotenv.
    if "OPENAI_API_KEY" not in os.environ:
         os.environ["OPENAI_API_KEY"] = os.environ("OPEN_API_KEY") # Replace with your actual key for testing

    input_article_content = "Scientists have discovered a new species of deep-sea fish with bioluminescent properties, expanding our understanding of marine life."
    
    # Example 1: Search among all articles
    print("--- Searching among all articles ---")
    similar_content, similar_article_data = RAG_find_similar_news_article(input_article_content)
    print(f"\nInput News Content: {input_article_content}")
    print(f"\nMost Similar News Article Content:\n{similar_content}")
    print(f"Details of Similar Article: {similar_article_data}")

    print("\n" + "="*50 + "\n")

    # Example 2: Search among a subset of articles (e.g., only news_1, news_3, news_6)
    print("--- Searching among a subset of articles (news_1, news_3, news_6) ---")
    subset_ids = ["news_1", "news_3", "news_6"]
    similar_content_subset, similar_article_data_subset = RAG_find_similar_news_article(
        input_article_content, subset_ids
    )
    print(f"\nInput News Content: {input_article_content}")
    print(f"\nMost Similar News Article Content (from subset):\n{similar_content_subset}")
    print(f"Details of Similar Article (from subset): {similar_article_data_subset}")

    print("\n" + "="*50 + "\n")
    
    # Example 3: A query that is very similar to an existing one
    input_article_content_2 = "A new artificial intelligence assistant was launched by a major tech firm, aiming to change how we do daily tasks."
    print("--- Searching for a very similar article ---")
    similar_content_2, similar_article_data_2 = RAG_find_similar_news_article(input_article_content_2)
    print(f"\nInput News Content: {input_article_content_2}")
    print(f"\nMost Similar News Article Content:\n{similar_content_2}")
    print(f"Details of Similar Article: {similar_article_data_2}")
```
