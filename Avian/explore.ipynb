```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Simulate delivery locations (latitude, longitude)
np.random.seed(42)
num_deliveries = 100
deliveries = pd.DataFrame({
    'latitude': np.random.uniform(37.5, 37.8, num_deliveries),
    'longitude': np.random.uniform(-122.5, -122.2, num_deliveries),
    'is_depot': False  # Mark all deliveries as non-depot
})

# Add depot (central hub)
depot = pd.DataFrame({
    'latitude': [37.65],  # Central location
    'longitude': [-122.35],
    'is_depot': True
})
df = pd.concat([depot, deliveries], ignore_index=True)

# Cluster using K-Means (include depot in data but exclude from clustering)
n_clusters = 4  # Number of drivers
kmeans = KMeans(n_clusters=n_clusters)
df['cluster'] = kmeans.fit_predict(df[['latitude', 'longitude']])

# Force depot to be part of all clusters (optional: use if drivers must return to depot between deliveries)
# df.loc[df['is_depot'], 'cluster'] = -1  # Uncomment to mark depot as separate

# Plot clusters
plt.figure(figsize=(10, 6))
for cluster in range(n_clusters):
    cluster_data = df[df['cluster'] == cluster]
    plt.scatter(cluster_data['longitude'], cluster_data['latitude'], 
               label=f'Cluster {cluster}', alpha=0.6)
# Highlight depot
plt.scatter(df[df['is_depot']]['longitude'], df[df['is_depot']]['latitude'], 
           c='red', marker='s', s=200, label='Depot')
plt.scatter(kmeans.cluster_centers_[:, 1], kmeans.cluster_centers_[:, 0], 
           c='black', marker='X', s=200, label='Centroids')
plt.title("Delivery Clusters with Depot")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend()
plt.show()

# Print cluster assignments
print("Deliveries per cluster:")
print(df[~df['is_depot']].groupby('cluster').size())
```

```python
from sklearn.cluster import KMeans
import numpy as np

# Sample input: coordinates, demands, and identical time windows
locations = np.array([
    [2, 3], [3, 4], [4, 5], [8, 8], [9, 10],
    [10, 9], [1, 1], [2, 2], [6, 6], [7, 7]
])
demands = [5, 10, 8, 7, 6, 4, 3, 2, 9, 6]
vehicle_capacity = 15
num_vehicles = 3

# Step 1: KMeans clustering by location
kmeans = KMeans(n_clusters=num_vehicles, random_state=0)
labels = kmeans.fit_predict(locations)

# Step 2: Enforce vehicle capacity in clusters
clusters = {i: {'locs': [], 'demand': 0} for i in range(num_vehicles)}
overflow = []

for idx, label in enumerate(labels):
    if clusters[label]['demand'] + demands[idx] <= vehicle_capacity:
        clusters[label]['locs'].append(idx)
        clusters[label]['demand'] += demands[idx]
    else:
        overflow.append(idx)

# Step 3: Try to assign overflow to any feasible cluster
for idx in overflow:
    placed = False
    for cluster in clusters.values():
        if cluster['demand'] + demands[idx] <= vehicle_capacity:
            cluster['locs'].append(idx)
            cluster['demand'] += demands[idx]
            placed = True
            break
    if not placed:
        print(f"⚠️ Customer {idx} could not be assigned due to capacity constraints.")

# Output result
for i, c in clusters.items():
    print(f"Cluster {i}: customers={c['locs']} | total_demand={c['demand']}")
```
